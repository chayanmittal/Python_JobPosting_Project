```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv(r'C:\Users\rsrcy\OneDrive\Desktop\ChayanPersonal\VSC\DA with Python\JobPosting_Project\data_jobs.csv')

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df_india= df[df['job_country']=='India'].copy()
df_india
```

```python
df_india['job_posted_month']=df_india['job_posted_date'].dt.strftime('%B')
```

```python
df_india_pivot = df_india.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_india_pivot.reset_index(inplace=True)
```

```python

df_india_pivot['month_num']=pd.to_datetime(df_india_pivot['job_posted_month'], format='%B').dt.month
```

```python
df_india_pivot.sort_values(by='month_num',inplace=True)
df_india_pivot.set_index('job_posted_month',inplace=True)
df_india_pivot.drop(columns='month_num',inplace=True)
df_india_pivot
```

```python
df_india_pivot.plot(kind='line')
```

```python
top_5 = df_india['job_title_short'].value_counts().head()
top_5=top_5.index.tolist()
```

```python
df_india_pivot[top_5].plot(kind='line')
plt.title('Monthly job posting for top 5 Data science roles')
plt.xlabel('2023')
plt.ylabel('Job counts')
plt.show()
```

```python
df_software = pd.read_csv(r'C:\Users\rsrcy\OneDrive\Desktop\ChayanPersonal\VSC\DA with Python\JobPosting_Project\software_jobs.csv',
                          index_col='job_posted_month')
df_software
```

```python
df_india_merged = df_india_pivot.merge(df_software,on='job_posted_month')
```

```python
top_6 = (
    df_india_merged
    .sum()
    .sort_values(ascending=False)
    .head(6)
    .index  #to get index of top 6
    .to_list()
)
top_6
```

```python
df_india_merged[top_6].plot(kind='line')
plt.title('Monthly job posting for top 6 roles')
plt.xlabel('2023')
plt.ylabel('Job counts')
plt.ylim(0,3000)
plt.legend()
plt.show()
```
